```python
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from bs4 import BeautifulSoup
import    urllib.request
import pandas as pd
import numpy as np
import time
import re


url = 'https://www.youtube.com/c/PeladoNerd/videos'

browser = webdriver.Chrome()
#browser = webdriver.Chrome(executable_path='C:/path/to/chromedriver.exe'

browser.get(url)

elements = browser.find_element_by_tag_name('html')

for i in range(50):
    elements.send_keys(Keys.END)
    time.sleep(2)


contador = 0

elements = elements.find_elements_by_id('video-title')
browser.minimize_window()
#print(elements)https://www.youtube.com/user/uneatlantico/videos
```

```python
tupla = []
data = []
contador = 0
for x in elements:
    vews = x.get_attribute("aria-label")
    try:
        if ((vews != 'Especial coronavirus: frenar la expansión by DW Español 12 minutes, 31 seconds No views') and (vews != 'Lola Arias, directora argentina de teatro by DW Español 2 months ago 26 minutes 665 views')):
            re.search(r'\bago \d\d|\bago \d,|seconds \d|second \d|hours \d|hour \d|minutes \d|minute \d|other \b',vews).group(0)
            text = re.search('(hour |seconds |second |hours |minute |minutes |ago )(?=\d,|\d\d,|\d\d\d,)(.*?) views', str(vews)).group(2)
            tupla.append(x.text)
            tupla.append(text)
            data.append(tupla)
#             print(tupla)
            tupla = []
            contador = contador + 1
    except:
        re.search(r'\bago \d\d|\bago \d,|seconds \d|second \d|hours \d|hour \d|minutes \d|minute \d|other \b',vews).group(0)
        text = re.search('(hour |seconds |second |hours |minute |minutes )(?=\d,|\d\d|\d\d\d)(.*?) views', str(vews)).group(2)
        tupla.append(x.text)
        tupla.append(text)
        data.append(tupla)
#         print(vews)
#         print(tupla)
        tupla = []
        contador = contador + 1
#print(contador)
```
